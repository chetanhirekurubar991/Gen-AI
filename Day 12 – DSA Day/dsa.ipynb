```
def lon_substring(string:str)->int:
    l=0
    r=1
    l_string=[]
    if not string:
        return 0
    elif len(string)==1:
        l_string.append(string[0])
        return len(l_string)
    else:
        l_string.append(string[l])
        while r<len(string):
            if string[r] not in l_string:
                l_string.append(string[r])
                r+=1
            else:
                l+=1
                r+=1
        if l>=len(l_string)-1:
            print("".join(l_string))
            return len(l_string)
        else:
            result=len(l_string)-l
            l_string=l_string[-result:]
            print("".join(l_string))
            return result

lon_substring("tmmzuxt")
```
