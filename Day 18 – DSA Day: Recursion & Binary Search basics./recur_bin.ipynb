#### Problem 1: Countdown (Recursion - Easy) üöÄ


```python
def countdown(n:int):
    if n==0:
        print("BlastOff!")
        return
    print(n)
    countdown(n-1)
countdown(5)
```

#### Problem 2: Sum of Array (Recursion - Easy/Medium) üî¢

```python
def sum_array(array):
    if not array:
        return 0
    return array[0]+sum_array(array[1:])
sum_array([1,2,3,4])
```

#### Problem 3: Binary Search (Iterative - Medium) üîç

```python
def binary_search(array,target):
    if not array:
        return -1
    else:
        l,r=0,(len(array)-1)
        while l<=r:
            mid=(l+r)//2
            if array[mid]==target:
                return mid
            elif array[mid]<target:
                l=mid+1
            else:
                r=mid-1
        return -1

arr = [1, 3, 5, 7, 9, 11, 13, 15, 17]
binary_search(arr,6)
```

#### Problem 4: Binary Search (Recursive - Medium) üîç

```python
def bin_search_recur(array,target,left,right):
    mid=(left+right)//2
    if not array:
        return -1
    elif array[mid]==target:
        return mid
    elif array[mid]<target:
        left=mid+1
    else:
        right=mid-1
    return bin_search_recur(array[left:right],target,left,right)
arr = [1, 3, 5, 7, 9, 11, 13, 15, 17]
bin_search_recur(arr, 6, 0, len(arr)-1)
```
