```
class Node:
    def __init__(self,val=0,next=None):
        self.val=val
        self.next=next
def removeNele(head,N):
    dummy=Node(0)
    dummy.next=head
    fast=slow=dummy
    for i in range(N+1):
        fast=fast.next()

    while fast:
        fast=fast.next
        slow=slow.next
    slow.next=slow.next.next
    return dummy.next
```
