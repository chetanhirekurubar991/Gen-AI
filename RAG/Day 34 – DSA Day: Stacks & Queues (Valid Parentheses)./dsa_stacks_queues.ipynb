#### Longest validate parenthesis

```python
def lon_vali_para(s:str):
    stack=[-1]
    length=0
    for i,char in enumerate(s):
        if char=="(":
            stack.append(i)
        else:
            stack.pop()
            if not stack:
                stack.append(i)
            else:
                length=max(length,i-stack[-1])
    return length
lon_vali_para("(())")
```

#### Min number of parentheses to remove to make string valid.

```python
def min_no_para(s):
    stack=[]
    count=0
    for i, char in enumerate(s):
        if char not in ['(',')']:
            continue
        else:
            if char =='(':
                stack.append(i)
            else:
                if not stack:
                    count+=1
                else:
                    stack.pop()
    total_counts=len(stack)+count
    return total_counts

min_no_para("))((")
```

#### Score of Parentheses


```python
def score_of_para(s):
    stack=[0]
    for char in s:
        if char == '(':
            stack.append(0)
        elif char ==')':
            if not stack:
                continue
            else:
                top=stack.pop()
                stack[-1]+=max(2*top,1)
        else:
            continue
    return stack
score_of_para('(()((())))')
```
